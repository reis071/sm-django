**# Aula: Criando um Projeto Simples em Django**

## **Objetivo**
Nesta aula, vamos criar um projeto Django básico do zero. O objetivo é entender a estrutura de um projeto Django e como criar uma aplicação simples com uma página inicial e um formulário para cadastrar informações no banco de dados.

## **Passo 1: Criar um Ambiente Virtual**
1. Crie um diretório para o projeto.
2. Dentro do diretório, crie um ambiente virtual.
3. Ative o ambiente virtual.
4. Instale o Django.

## **Passo 2: Criar o Projeto Django**
1. Use o comando para criar um novo projeto Django.
2. Entre no diretório do projeto.
3. Execute o servidor para testar se o Django foi instalado corretamente.

## **Passo 3: Criar um Aplicativo Django**
1. No diretório do projeto, crie um aplicativo Django.
2. Registre o aplicativo no arquivo `settings.py`.

## **Passo 4: Configurar as URLs**
1. No arquivo `urls.py` do projeto, inclua a URL do aplicativo.
2. No aplicativo, crie um arquivo `urls.py`.
3. Defina a rota para a página inicial.

## **Passo 5: Criar uma View Simples**
1. No arquivo `views.py` do aplicativo, crie uma função que retorne uma resposta simples.
2. Associe a função à URL no `urls.py` do aplicativo.

## **Passo 6: Criar um Template HTML**
1. Crie uma pasta `templates/` dentro do aplicativo.
2. Dentro dessa pasta, crie um arquivo HTML para a página inicial.
3. Modifique a view para renderizar o template.

## **Passo 7: Trabalhando com Models**
1. No arquivo `models.py`, crie um modelo para armazenar informações.
2. Faça as migrações para aplicar o modelo ao banco de dados.

## **Passo 8: Criar um Formulário**
1. No arquivo `forms.py`, crie um formulário baseado no modelo.
2. No `views.py`, crie uma view para lidar com o formulário.
3. No `urls.py`, adicione uma rota para o formulário.
4. No diretório `templates/`, crie um template para o formulário.
5. Teste o envio de informações pelo formulário.

## **Passo 9: Visualizar Dados**
1. No `views.py`, crie uma view para listar os dados do banco.
2. No `urls.py`, adicione uma rota para listar os dados.
3. No diretório `templates/`, crie um template para exibir os dados.

## **Passo 10: Melhorando a Navegação**
1. Adicione links de navegação entre as páginas.
2. Modifique o template base para melhorar a interface.

## **Passo 11: Testar e Refatorar**
1. Teste todas as funcionalidades.
2. Faça ajustes conforme necessário.
3. Adicione estilos básicos com CSS.

## **Conclusão**
Ao final dessa aula, teremos um projeto Django funcional com uma página inicial, um formulário de cadastro e uma listagem de dados do banco. Esse é o primeiro passo para entender o funcionamento do Django e como criar aplicações mais complexas.
